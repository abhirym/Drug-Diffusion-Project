## Problem Statement

In a clinical setting, a physician sees outcomes — did the drug work or not? 
What they cannot easily observe is *why*: how the drug moved through tissue, 
where it was lost to binding or degradation, and at what depth the concentration 
fell below the therapeutic threshold.

This gap between clinical observation and molecular reality is where computational 
simulation is most valuable. By modeling diffusion from first principles, we can 
make the invisible visible — tracking drug concentration across tissue depth and 
time under conditions that would be difficult or impossible to isolate experimentally.

This project uses Fick's second law of diffusion as its mathematical foundation, 
progressively adding biological complexity across five experiments to build toward 
a physically realistic model of drug behavior in tissue.

## Scientific Background

### Fick's Second Law of Diffusion
Fick's second law describes how the concentration of a substance changes over 
time as it diffuses through a medium:

∂C/∂t = D · (∂²C/∂x²)

Where:
- `C` is the concentration of the diffusing substance (drug)
- `t` is time
- `x` is position (depth into tissue)
- `D` is the diffusion coefficient — a material property describing how easily 
the substance moves through the medium

The analytical solution for diffusion from a point source is a Gaussian 
concentration profile:

C(x, t) = (1 / √(4πDt)) · exp(−x² / 4Dt)

This is the foundation of Experiments 1 through 3.

### Diffusion Coefficient (D)
D is not a fixed constant — it varies with the drug's molecular size, shape, 
and charge, as well as the physical properties of the tissue it moves through. 
Larger molecules diffuse more slowly. Denser or more structured tissue resists 
diffusion. Experiment 3 explores how sensitive drug penetration is to changes in D.

### Layered Tissue
Real biological tissue is not uniform. Skin, for example, consists of multiple 
layers — epidermis, dermis, subcutaneous tissue — each with different structural 
properties and therefore different diffusion coefficients. A drug crossing these 
boundaries encounters changing resistance. Experiment 4 introduces a two-layer 
tissue model to capture this discontinuity.

### Molecular Binding
As a drug diffuses through tissue, it encounters proteins, receptors, and other 
molecules it can bind to. Bound drug is no longer free to diffuse — it is 
effectively removed from the concentration gradient. This reduces both the 
magnitude and the penetration depth of free drug across the tissue. Experiment 5 
models this effect and its clinical implications.

## A Note on Scientific Transparency

The binding model in Experiment 5 is currently implemented as a first-order 
decay term applied to the analytical diffusion solution:

C_bind = exp(−x² / 4Dt) · exp(−k_bind · t)

This is a valid approximation of uniform drug degradation or nonspecific uptake, 
but it does not accurately capture true reversible molecular binding, which is:
- **Concentration-dependent** — binding rate scales with how much free drug is present
- **Saturable** — binding sites get used up as drug occupies them
- **Reversible** — bound drug can unbind and re-enter the free pool

This limitation is documented here deliberately. The next version of Experiment 5 
will replace this term with a coupled PDE system modeling reversible binding kinetics:

∂C/∂t       = D(∂²C/∂x²) − k_on·C·B + k_off·C_bound  
∂C_bound/∂t = k_on·C·B − k_off·C_bound  
∂B/∂t       = −k_on·C·B + k_off·C_bound  

Where `k_on` and `k_off` will be drawn from pharmacokinetic literature for a 
specific drug class rather than set arbitrarily.
