# Experiment 3: Varying the Diffusion Coefficient
**Drug Modeled:** Lidocaine  
**Condition Tested:** Sensitivity of concentration profile to changes in D  
**Research Question:** How does tissue density and drug formulation affect 
Lidocaine penetration depth, as modeled by varying the diffusion coefficient?

## Background

The diffusion coefficient D is not a fixed biological constant — it varies 
based on two key factors:

**Tissue properties:** Denser, more structured tissue (scar tissue, inflamed 
tissue, connective tissue) resists diffusion more than loose soft tissue. 
This means D is effectively lower in denser tissue, reducing how far 
Lidocaine can penetrate.

**Drug formulation:** Lidocaine can be delivered in different formulations 
that alter its effective diffusivity. Adding vasoconstrictors like epinephrine, 
changing pH, or using different carrier agents all affect how quickly the 
drug moves through tissue.

This experiment uses the validated baseline D = 1.25 × 10⁻⁸ cm²/s from 
literature as the reference value, then models the effect of halving and 
doubling D to represent denser and more permeable tissue respectively.

All other parameters are held constant at t = 10 minutes to isolate the 
effect of D alone.

```python
import numpy as np
import matplotlib.pyplot as plt
import os

os.makedirs('figures', exist_ok=True)

# ── Parameters ────────────────────────────────────────────────────────────────

# Baseline diffusion coefficient — Lidocaine through biological membrane
# Source: University of Minnesota, NIH
D_baseline = 1.25e-8     # cm²/s

# Varying D to model different tissue densities and formulations
D_values = [
    D_baseline * 0.25,   # Dense tissue (e.g. scar tissue, inflamed tissue)
    D_baseline * 0.5,    # Moderately dense tissue
    D_baseline,          # Baseline — validated literature value
    D_baseline * 2,      # More permeable tissue
    D_baseline * 4,      # Highly permeable tissue
]

D_labels = [
    '0.25× D — Dense tissue',
    '0.5× D — Moderately dense',
    '1× D — Baseline (literature)',
    '2× D — More permeable',
    '4× D — Highly permeable',
]

# Initial surface concentration — 2% Lidocaine solution
C0 = 20.0                # mg/mL

# Tissue depth range
x = np.linspace(0, 0.15, 1000)   # cm (0 to 1.5mm)

# Fixed time point — 10 minutes after administration
t = 600                  # seconds

# Color palette
colors = ['#d62728', '#ff7f0e', '#2ca02c', '#1f77b4', '#9467bd']
```

```python
# ── Concentration Profiles Across D Values ────────────────────────────────────

profiles = []

for D in D_values:
    normalization = 1 / np.sqrt(4 * np.pi * D * t)
    C = C0 * normalization * np.exp(-(x**2) / (4 * D * t))
    profiles.append(C)
```

```python
# ── Plot ──────────────────────────────────────────────────────────────────────

fig, ax = plt.subplots(figsize=(9, 5))

for i, (C, label) in enumerate(zip(profiles, D_labels)):
    linestyle = '--' if i == 2 else '-'  # dashed for baseline
    ax.plot(x * 10, C, color=colors[i], linewidth=2,
            linestyle=linestyle, label=label)

ax.set_title('Effect of Diffusion Coefficient on Lidocaine Penetration Depth',
             fontsize=13, fontweight='bold')
ax.set_xlabel('Depth into Tissue (mm)', fontsize=11)
ax.set_ylabel('Concentration (mg/mL)', fontsize=11)
ax.legend(fontsize=9, title='Tissue / Formulation Condition')
ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig('figures/exp03_D_variation.png', dpi=150)
plt.show()
```

## Observations

- Higher D values produce flatter, wider concentration profiles — drug 
  penetrates deeper but surface concentration is lower
- Lower D values produce steeper, narrower profiles — drug stays concentrated 
  near the surface but fails to penetrate deeper tissue
- Small changes in D produce disproportionately large changes in penetration 
  depth — the relationship is nonlinear
- The baseline literature value sits in the middle of the range, confirming 
  it represents typical soft tissue conditions

## Interpretation

The sensitivity of penetration depth to D has direct clinical implications. 
Patients with scar tissue, chronic inflammation, or fibrosis at the injection 
site effectively have a lower local D — meaning standard Lidocaine doses may 
fail to reach the target nerve depth even when correctly administered.

Conversely, highly vascularized or edematous tissue may have a higher effective 
D, causing drug to spread more rapidly but at lower local concentrations — 
potentially reducing anesthetic duration.

This experiment highlights why a one-size-fits-all dosing approach is 
insufficient and why computational modeling of patient-specific tissue 
properties could improve anesthetic outcomes.
